### Actividad 1: 
Suma de todos los elementos de una lista, utilizando bucles repetitivos (for, while).

```python

# todo:

numbers = [14, 23, 34, 48, 85]

suma = 0 # inicializamos la variable suma en 0

# bucle for para sumar los numeros de la lista
for i in numbers:
    suma = suma + i

print(suma)
```

### Actividad 4: 
Invertir una lista sin usar reverse() ni [::-1]

```python

# todo:

lists_two = [randint(1, 50) for _ in range(8)]
print("Lista original:", lists_two)

# funcion para revertir una lista
def revertir_sin_reverse(lista):
    for i in range(len(lista) // 2):  # Recorremos la lista mitad de la lista
        lista[i], lista[-i - 1] = lista[-i - 1], lista[i]  # Cambiamos los elementos de la lista
    return lista

revertir_sin_reverse(lists_two)
print("Lista invertida:", lists_two)
```

### Actividad 5:
Sumar los elementos de dos listas posición por posición, utilizando solo bucles repetititivos (for,while).

### Actividad 6:

Escribir un programa que almacene el abecedario en una lista, elimine de la lista las letras que ocupen posiciones múltiplos de 3, y muestre por pantalla la lista resultante.

```python

# todo:

abecedario = ['a','b','c','d','e','f','g','h','i','j','k','l','m','n','o','p','q','r','s','t','u','v','w','x','y','z']

# funcion que elimina los numeros multiplos de 3
def eliminar_multiplos_de_3(lista):
    return [letra for indice, letra in enumerate(lista) if indice % 3 != 0]

# se guarda el resultado en la variable abecedario_eliminado    
abecedario_eliminado = eliminar_multiplos_de_3(abecedario)

print(abecedario_eliminado)
```

### Actividad 7:

Escribir un programa que pida al usuario una palabra y muestre por pantalla si es un palíndromo (una palabra que se escribe de la misma forma al derecho y al reves).

```python

# todo:

palabra = input("Ingresa una palabra: ").lower()  # se pide la palabra y se convierte a minúsculas

if palabra == palabra[::-1]:  # Compara la palabra original con su versión invertida
    print(f"'{palabra}' es un palíndromo.")
else:
    print(f"'{palabra}' no es un palíndromo.")
```

### Actividad 10:

Escribir un programa que cree 2 matrices y realice su multiplicacion.

```python

# todo:

def crear_matriz(filas, columnas):
    # Se crea la matriz con los valores introducidos por el usuario
    return [[float(input(f"Elemento [{i+1}][{j+1}]: ")) 
             for j in range(columnas)] for i in range(filas)]

def multiplicar_matrices(A, B):
    # Verifica si las matrices son compatibles para multiplicación
    if len(A[0]) != len(B):
        print("Error: Las matrices no son compatibles para multiplicación")
        return None
    
    # Crea la matriz resultado
    resultado = [[0 for _ in range(len(B[0]))] for _ in range(len(A))]
    
    # Multiplicación
    for i in range(len(A)):
        for j in range(len(B[0])):
            for k in range(len(B)):
                resultado[i][j] += A[i][k] * B[k][j]
    return resultado

# se muestran las matrices y el resultado de la multiplicación
print("MATRIZ A")
filasA = int(input("Filas de A: "))
columnasA = int(input("Columnas de A: "))
A = crear_matriz(filasA, columnasA)

print("\nMATRIZ B")
filasB = int(input("Filas de B: "))
columnasB = int(input("Columnas de B: "))
B = crear_matriz(filasB, columnasB)

# Calcular y mostrar resultado
resultado = multiplicar_matrices(A, B)
if resultado:
    print("\nResultado A x B:")
    for fila in resultado:
        print(fila)
```
